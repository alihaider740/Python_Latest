```python
class car:
    def __init__(self,make,model,year,ownerInfo):
        self.make = make
        self.model = model
        self.year = year
        self.fuel = 0
        self.mileage = 0
        self.owner = ownerInfo
    def fuelUp(self, liters):
        self.fuel += liters
        print(f"Filling up {liters} liters of fuel.")

    def getCurrentFuel(self):
        return self.fuel

    def getOwnerInfo(self):
        return self.owner

    def getCurrentMileage(self):
        return self.mileage 

    def drive(self, distance):
        fuel_consumed = distance / 5
        if fuel_consumed > self.fuel:
            print("Not enough fuel to drive the distance.")
            return
        self.mileage += distance 
        self.fuel -= fuel_consumed
        print(f"Driving {distance} km.")

my_car = car("KIA","Picanto",2021,{
    "name" : "Haider"
})


my_car.fuelUp(30)
my_car.drive(20)


```
